# Q2: Show Bar Graphs representing the Departure Time & Arrival Time

```python
df.head()
```

```python
# Showing the departure Time for the flight
df['departure_time'].value_counts()
```

```python
# Showing the Arrival Time for the flight
df['arrival_time'].value_counts()
```

```python
# Showing the Departure Time & Arrival Time for the flights with their counts
plt.figure(figsize = (15, 4))
plt.subplot(1,2,1)
plt.bar( df['departure_time'].value_counts().index, df['departure_time'].value_counts().values, color = ['r', 'b'])
plt.title("Departure Time")
plt.xlabel("D Time")
plt.ylabel("D Freq")
plt.show()
```

```python
plt.figure(figsize = (16, 4))
plt.subplot(1,2,1)
plt.bar(df['arrival_time'].value_counts().index, df['arrival_time'].value_counts().values, color = ['g', 'y'])
plt.title("Arrival Time")
plt.xlabel("A Time")
plt.ylabel("A Freq")
plt.show()
```

# Q: 4 Does price varies with airlines ?

```python
df.head()
```

```python
# Grouping the airlines and checking their mean price

df.groupby("airline")['price'].mean()
```

```python
# Drawing a Categorical plot showing the mean Ticket price for each Airline
import warnings
warnings.filterwarnings('ignore')
sns.catplot(x='airline', y='price', kind='bar', palette='rocket', data=df, hue ='class' )
plt.show()
```

# Q:5 Does ticket price change based on the departure time and arrival time
